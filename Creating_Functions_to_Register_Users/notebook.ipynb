```python
def validate_user(name, email, password):
    if all([validate_name(name), validate_email(email), validate_password(password)]):
        return True
    else:
        raise ValueError('Something wrong with either name, email, or password')
        
def register_user(name, email, password):
    if validate_user(name, email, password):
        return {'name': name, 'email': email, 'password': password}
    else:
        return 
    
Name = 'John Doe'
Email = 'john.doe@example.com'
Password = 'P@ssw0rd123!'

print(register_user(Name, Email, Password))
```
