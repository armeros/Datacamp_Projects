```python
# Loading in required libraries
import pandas as pd, seaborn as sns, numpy as np, matplotlib.pyplot as plt

nobel = pd.read_csv('data/nobel.csv')
nobel.head(3)
```

### What is the *most commonly awarded gender and birth country*?

```python
display(nobel['sex'].value_counts())
display(nobel['birth_country'].value_counts().head(3))
top_country = nobel['birth_country'].value_counts().index[0]
top_gender = nobel['sex'].value_counts().index[0]
print(f'Top country is {top_country} and top_gender is {top_gender}.')
```

### Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?


```python
def decade(year):
    if 1900 <= year < 1910:
        return 1900
    elif 1910 <= year < 1920:
        return 1910
    elif 1920 <= year < 1930:
        return 1920
    elif 1930 <= year < 1940:
        return 1930
    elif 1940 <= year < 1950:
        return 1940
    elif 1950 <= year < 1960:
        return 1950
    elif 1960 <= year < 1970:
        return 1960
    elif 1970 <= year < 1980:
        return 1970
    elif 1980 <= year < 1990:
        return 1980
    elif 1990 <= year < 2000:
        return 1990
    elif 2000 <= year < 2010:
        return 2000
    elif 2010 <= year < 2020:
        return 2010
    else:
        return 2020
        
nobel['decade'] = nobel['year'].apply(decade)
total_winners = nobel.groupby('decade')['prize'].count()
us_nobel = nobel[nobel['birth_country'].isin(['USA', 'United States of America'])]
us_winners = us_nobel.groupby('decade')['prize'].count()
ratio = us_winners / total_winners
ratio.sort_values(ascending=False).head()
max_decade_usa = ratio.idxmax()
print(f'The decade that had the highest ratio of US-born Nobel Prize winners to total winners in all categories is {max_decade_usa}\'s.')
```

### Which decade and Nobel Prize category combination had the highest proportion of female laureates?

```python
total_winners = nobel.groupby(['decade', 'category'])['year'].count()
female_nobel = nobel[nobel['sex']=='Female']
female_winners = female_nobel.groupby(['decade', 'category'])['year'].count()
ratio = female_winners/ total_winners
max_female_dict = {ratio.idxmax()[0]: ratio.idxmax()[1]}
print(f'In {ratio.idxmax()[0]}, {ratio.idxmax()[1]} had the highest proportion of female winners.')

```

### Who was the first woman to receive a Nobel Prize, and in what category?

```python
for i, row in nobel.iterrows():
    if row['sex'] == 'Female':
        first_woman_name,  first_woman_category = row['full_name'], row['category']
        break
print(f'The first woman to receive a Nobel Prize is {first_woman_name} in {first_woman_category}.')
```
